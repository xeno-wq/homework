```python
def analyze_students(data):
    res={}
    words=[]
    first={}
    count_words=[]
    second={}
    for x in data:
        for y in x['comments']:
            if y not in words:
                words.append(y)
            count_words.append(y)
    
    for x in count_words:
        if x not in second:
            second[x]=0
        second[x]+=1    
    for x in words:
        if x not in res:
            res[x]=[]
    for x in data:
        first[x['name']]=x['comments'] 
               
    for x,y in first.items():
        for i in y:
            if i not in res:
                res[i]=[]
            res[i].append(x)
    for x,y in res.items():
        z=[]
        for i in y:
            if i not in z:
                z.append(i)
            res[x]=z   
    for x,y in second.items():
        for i in res.items():
            if x==i[0]:
                res[i[0]].append(y)
    res=dict(filter(lambda x:len(x[1])>=3,res.items()))
    res=dict(sorted(res.items(),key=lambda x:(-len(x[1]),x[0])))
    return res
data=[
    {'name':'Yamal','Grades':[9,8,7,9],'comments':['Excellent','Very good','Good','Excellent']},
    {'name':'Doue','Grades':[9,5,7,7],'comments':['Excellent','bad','Good','Good']},
    {'name':'Cubarsi','Grades':[8,8,6,5],'comments':['Very good','Very good','Good','bad']},
    {'name':'Gul2r','Grades':[8,7,6,5],'comments':['Very good','Good','Not bad','Bad']}
]
analyze_students(data)
```
