```python
def analyze(data):
    res={}
    m=0
    b=0
    for x,y in data.items():
        mean=0
        c=0
        for i in y:
            mean+=i
            m+=i
            b+=1
            c+=1
        res[x]=mean/c
        
    max_q=0
    best={}
    min_q=res['Ali']
    worst={}
    
    
    
    for x,y in res.items():
        if y>max_q:
            max_q=y
        if y<min_q:
            min_q=y
    
    
    
    for x in res.items():
        if x[1]==max_q:
            best[x[0]]=x[1]    
        if x[1]==min_q:
            worst[x[0]]=x[1] 
      
    return res,best,worst,m/b

data={
 "Ali": [80, 90, 75],
 "Dana": [100, 95, 92],
 "Oleg": [60, 70, 65]
}
mean,best,worst,mean=analyze(data)
print(mean)
print(best)
print(worst)
print(mean)
```

```python
def analyze_students(data):
    res=[]
    final={}
    second=[]
    for student in data:
        if student['name'].isalpha():
            student['name']=student['name'].title()
            res.append(student)
    first=[]
    def summa(x):
        summa=0
        while x!=0:
            summa+=x%10
            x//=10
        return summa
    
    for d in res:
        for x,y in d.items():
            if x=='comments':
                for i in y:
                    second.append(i)
    sc={}
    for x in second:
        if x not in sc:
            sc[x]=0
        sc[x]+=1
    
    
    for x in res:
        new=list(map(lambda x:summa(x) if x%2!=0 and x<10 else x**2 if x%2==0 and x>=10 else x,filter(lambda x:x>0,x['Grades'])))
        x['Grades']=new
        first.append(x) 
        x['processed_grades']=x.pop('Grades')                
    for student in data:
        if 'comments' in student:
            student.pop('comments')
    third=set()
    for x in second:
        a='aeuioAEUIO'
        for y in x:
            if y in a:
                third.add(y)   
    students_by_avg = [
        s['name'] 
        for s in sorted(res, key=lambda x: sum(x['processed_grades'])/len(x['processed_grades']), reverse=True)
    ]
    fifth={}
    for d in res:
        for x,y in d.items():
            if x=='name':
                if x not in fifth:
                    fifth[len(y)]=[]
                fifth[len(y)].append(y)
    fifth=dict(sorted(fifth.items(),key=lambda x:-x[0]))
    
    
    
    final['Students']=first
    final['word_counts']=sc
    final['all_vowels']=third
    final['students_by_avg']=students_by_avg
    final['students_by_name_length']=fifth
    return final      







data=[
    {'name':'Yamal','Grades':[9,8,7,9],'comments':['Excellent','Very good','Good','Excellent']},
    {'name':'Doue','Grades':[9,5,7,7],'comments':['Excellent','bad','Good','Good']},
    {'name':'Cubarsi','Grades':[8,8,6,5],'comments':['Very good','Very good','Good','bad']},
    {'name':'Gul2r','Grades':[8,7,6,5],'comments':['Very good','Good','Not bad','Bad']}
]
analyze_students(data)
```
