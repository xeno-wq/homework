```python
import csv
with open('synthetic_mobile_sales_2025.csv','r') as sm:
    lst=csv.DictReader(sm)
    count_sales=0
    total_revenue=0
    brands={}
    revenue_by_month={}
    for line in lst:
        count_sales+=1
        total_revenue+=int(line['Revenue_USD'])
        if line['Brand'] not in brands:
            brands[line['Brand']]=0
        brands[line['Brand']]+=int(line['Revenue_USD'])
        if line['Sale_Month'] not in revenue_by_month:
            revenue_by_month[line['Sale_Month']]=0
        revenue_by_month[line['Sale_Month']]+=int(line['Revenue_USD'])
avg=sum(list(brands.values()))/len(brands)
brands_above_average=dict(filter(lambda x:x[1]>avg,brands.items()))
the_most_active_brand=sorted(brands_above_average.items(),key=lambda x:-x[1])
top_3_months=sorted(revenue_by_month.items(),key=lambda x:-x[1])[:3]
with open('report.txt','w',encoding='UTF-8') as rp:
    rp.write(f'Общее количество продаж:{count_sales}\n')
    rp.write(f'Общая выручка:{total_revenue} USD \n')
    rp.write(f'Средний чек:{total_revenue/count_sales} USD \n')
    rp.write(f'Продажи по брендам:{brands}\n')
    rp.write(f'Бренды выше среднего:{brands_above_average}\n')
    rp.write(f'Самый активный бренд:{the_most_active_brand[0][0]}\n')
    rp.write(f'Выручка по месяцам:{revenue_by_month}\n')
    rp.write(f'Топ-3 месяца:{top_3_months}\n')
    rp.write(f'Самый прибыльный месяц:{top_3_months[0][0]}\n')
```
