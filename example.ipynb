```python
def pairwise_intersections(sets_list):                 #2.32
    if len(sets_list)<2:
        return []
    res=[]
    for i in range(len(sets_list)-1):
        x=sets_list[i]
        y=sets_list[i+1]
        
        res.append(x&y)
    return res
sets_list = [
    {1, 2, 3, 4},
    {3, 4, 5, 6},
    {4, 6, 7, 8}
]
pairwise_intersections(sets_list)
```

```python
def multi_symmetric_difference(sets_list):          #2.38
    res=[]
    for x in sets_list:
        for j in x:
            res.append(j)
    r=set()
    for x in res:
        if res.count(x)==1:
            r.add(x)
    return r
sets_list=[{1,2,3},{2,3,4},{3,4,5}]
multi_symmetric_difference((sets_list))
```

```python
def analyze_dict_keys(d):            #2.40
    res=set()
    for x in d.keys():
        if x.isalpha():
            for y in x:
                if x.count(y)==1:
                    res.add(y)
    return res
d={
    'apple':2,
    'banana':4,
    'coach':7
}
analyze_dict_keys(d)
```

```python
def analyze_students(data):             #3.1
    res=[]
    final={}
    second=[]
    for student in data:
        if student['name'].isalpha():
            student['name']=student['name'].title()
            res.append(student)
    first=[]
    def summa(x):
        summa=0
        while x!=0:
            summa+=x%10
            x//=10
        return summa
    
    for d in res:
        for x,y in d.items():
            if x=='comments':
                for i in y:
                    second.append(i)
    sc={}
    for x in second:
        if x not in sc:
            sc[x]=0
        sc[x]+=1
    
    
    for x in res:
        new=list(map(lambda x:summa(x) if x%2!=0 and x<10 else x**2 if x%2==0 and x>=10 else x,filter(lambda x:x>0,x['Grades'])))
        x['Grades']=new
        first.append(x) 
        x['processed_grades']=x.pop('Grades')                
    for student in data:
        if 'comments' in student:
            student.pop('comments')
    third=set()
    for x in second:
        a='aeuioAEUIO'
        for y in x:
            if y in a:
                third.add(y)   
    students_by_avg = [
        s['name'] 
        for s in sorted(res, key=lambda x: sum(x['processed_grades'])/len(x['processed_grades']), reverse=True)
    ]
    fifth={}
    for d in res:
        for x,y in d.items():
            if x=='name':
                if x not in fifth:
                    fifth[len(y)]=[]
                fifth[len(y)].append(y)
    fifth=sorted(fifth.items(),key=lambda x:-x[0])
    
    
    
    final['Students']=first
    final['word_counts']=sc
    final['all_vowels']=third
    final['students_by_avg']=students_by_avg
    final['students_by_name_length']=fifth
    return final      







data=[
    {'name':'Yamal','Grades':[9,8,7,9],'comments':['Excellent','Very good','Good','Excellent']},
    {'name':'Doue','Grades':[9,5,7,7],'comments':['Excellent','bad','Good','Good']},
    {'name':'Cubarsi','Grades':[8,8,6,5],'comments':['Very good','Very good','Good','bad']},
    {'name':'Gul2r','Grades':[8,7,6,5],'comments':['Very good','Good','Not bad','Bad']}
]
analyze_students(data)
```
