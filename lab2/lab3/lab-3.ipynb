1. Лямбда-функции (6 задач)


```python
function=lambda x:'положительное' if x>0 else  'отрицательное' if x<0 else 'ноль'          #1.1
print(function(5))
```

```python
words = ["арбуз", "кот", "машина", "дом", "ананас"]            #1.2
function=sorted(words,key=lambda x:(len(x),x))
print(function)
```

```python
numbers = [5, 12, 7, 20, 33, 8]                       #1.3
numbers=list(filter(lambda x:x%2==0 and x>10,numbers))
print(numbers)
```

```python
numbers = [1, 2, 3, 4, 5, 6]                                           #1.4
numbers=list(map(lambda x:x**2 if x%2==0 else x*3,numbers))
print(numbers)
```

```python
compare=lambda a,b:'a больше' if a>b else 'b больше' if b>a else 'равны'        #1.5

print(compare(10, 7))
```

```python
numbers = [0, -3, 5, -7, 8]                                               #1.6
numbers=list(map(lambda x:'положительное' if x>0 else 'отрицательное' if x<0 else 'ноль',numbers))
print(numbers)
```

2. Генераторы (4 задачи)

```python
n=int(input())                            #2.1
even_numbers=[x**2 for x in range(1,n+1)]
even_numbers=list(map(lambda x:'кратно 4' if x%4==0 else x,even_numbers))
print(even_numbers)
```
