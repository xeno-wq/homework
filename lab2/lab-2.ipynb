```python
with (open('shop_logs.txt','r') as sh):                    #1.1
    selling=0
    people=set()
    summa=0
    the_most_active_person={}
    for line in sh:
        parts=line.strip().split(';')
        user=parts[1]
        active=parts[2]
        people.add(user)
        if active.upper()=='BUY':
            selling+=1
            summa+=int(parts[3])

            if user not in the_most_active_person:
                the_most_active_person[user]=[]
            the_most_active_person[user].append(int(parts[3]))
the_most_active_person=dict(map(lambda x:(x[0],sum(x[1])),the_most_active_person.items()))
the_most_active_person=sorted(the_most_active_person.items(),key=lambda x:x[1],reverse=True)
with open('report.txt','w',encoding='utf-8') as rp:
    rp.write(f'Уникальных пользователей:{people}\n')
    rp.write(f'Всего покупок:{selling}\n')
    rp.write(f'Общая сумма:{summa}\n')
    rp.write(f'Самый активный покупатель:{the_most_active_person[0]}\n')
    rp.write(f'Средний чек:{summa/selling}\n')
with open('report.txt','r',encoding='utf-8') as rp:
    print(rp.read())
```

```python
import csv                            #2.1
with open('employees.csv','r') as emp:
    reader=csv.DictReader(emp)
    summa_salary=0
    names=[]
    dep_salary={}
    empl_salary={}
    for line in reader:
        names.append(line['name'])
        summa_salary+=int(line['salary'])
        if  line['department'] not in dep_salary:
            dep_salary[line['department']]=[]
        dep_salary[line['department']].append(int(line['salary']))
        if line['name'] not in empl_salary:
            empl_salary[line['name']]=int(line['salary'])

mx_salary=sorted(empl_salary.items(),key=lambda x:-x[1])[0]
dep_salary=dict(map(lambda x:(x[0],sum(x[1])/len(x[1])),dep_salary.items()))
dep_salary=sorted(dep_salary.items(),key=lambda x:-x[1])
empl_salary=dict(filter(lambda x:x[1]>summa_salary/len(names),empl_salary.items()))
print('среднюю зарплату:',summa_salary/len(names))
print('среднюю зарплату по каждому отделу',dep_salary)
print('отдел с самой высокой средней зарплатой',dep_salary[0])
print('самого высокооплачиваемого сотрудника',mx_salary)
print('список сотрудников с зарплатой выше средней',empl_salary)
with open('high_salary.csv','w',encoding='utf-8') as hs:
    hs.write(f'список сотрудников с зарплатой выше средней:{empl_salary}\n')
with open('high_salary.csv','r',encoding='utf-8') as hs:
    print(hs.read())
```

```python
import json                                     #3.1
with open('orders.json','r') as ord:
    reader=json.load(ord)
    sell_items=[]
    summa_orders=0
    cus_items={}
    cus_total={}
    for line in reader:
        summa_orders+=line['total']
        sell_items+=line['items']
        cus_items[str(line['order_id'])+'_'+line['user']]=line['items']
        cus_total[str(line['order_id'])+'_'+line['user']]=line['total']

count_item_per_user=dict(map(lambda x:(x[0],len(x[1])),cus_items.items()))
print('общую сумму всех заказов:',summa_orders)
print('сколько заказов сделал каждый пользователь',len(sell_items))
print('сколько заказов сделал каждый пользователь',count_item_per_user)
print('пользователя с самым дорогим заказом',sorted(cus_total.items(),key=lambda x:-x[1])[0])
print('самый популярный товар:',sorted(sell_items,key=lambda x:-sell_items.count(x))[0])
with open('summary.json','w',encoding='utf-8') as summ:
    summ.write(f"total_revenue:{summa_orders}\n")
    summ.write(f"top_user:{sorted(cus_total.items(),key=lambda x:-x[1])[0]}\n")
    summ.write(f"most_popular_item:{sorted(sell_items,key=lambda x:-sell_items.count(x))[0]}\n")
    summ.write(f"total_orders:{len(sell_items)}\n")
with open('summary.json','r',encoding='utf-8') as summ:
    print(summ.read())
```

```python
import csv                                #4.1
with open('transactions.csv','r') as tr:
    reader=csv.DictReader(tr)
    suspicious_transactions={}
    users={}
    summa=0
    for line in reader:
        if int(line['amount'])>500000:
            suspicious_transactions[line['user_id']]=int(line['amount'])
            summa+=int(line['amount'])
        if line['user_id'] not in users:
            users[line['user_id']]=[]
        users[line['user_id']].append(int(line['amount']))
all_users=list(map(lambda x:x[0],users.items()))
suspicious_users=dict(filter(lambda x:len(x[1])>3,users.items()))
summa_sus=list(map(lambda x:x[1],suspicious_users))


with open('fraud_report.txt','w',encoding='utf-8') as fr:
    fr.write(f'Подозрительных транзакций:{suspicious_transactions}\n')
    fr.write(f'Подозрительных пользователей:{suspicious_users}\n')
    fr.write(f'Список пользователей:{users}\n')
    fr.write(f'Общая сумма подозрительных операций:{sum(summa_sus)+summa}\n')
with open('fraud_report.txt','r',encoding='utf-8') as fr:
    print(fr.read())
with open('fraud_users.json ','w',encoding='utf-8') as fr:
    fr.write(f'Подозрительных пользователей:{suspicious_users}\n')
with open('fraud_users.json ','r',encoding='utf-8') as fr:
    print(fr.read())
```
